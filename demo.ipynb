```python
from llm_from_scratch.gpt.model.custom_gpt import CustomGPTwithRoPE
from llm_from_scratch.tokenizer.bpe_tokenizer import BytelevelBPE
```

```python
tokenizer_state_path = './tokenizer_state_sft'
ckpt_path = './SFT_GPTwithRoPE_best_val.pth'

tokenizer = BytelevelBPE()
tokenizer.load(tokenizer_state_path)
```

```python
import torch

ckpt = torch.load(ckpt_path)
GPT_CFG = ckpt['GPT_CFG'] 
```

```python
model = CustomGPTwithRoPE(cfg=GPT_CFG)
model.to('cuda:0')
model.load_state_dict(ckpt['model_state_dict'])
```

```python
len(tokenizer.vocab)
```

```python
input_query = "Hi!"
input_tokens = tokenizer.encode(f"<user>{input_query}<assistant>")

output_tokens = model.generate(input_tokens, 2048, 100, eos_token_id=tokenizer.special_token_dict['<eos>'], temperature=0.9, top_k=64)
print(tokenizer.decode(output_tokens).split('<assistant>')[-1])
```
